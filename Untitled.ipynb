```python
import pandas as pd
import csv
```

```python
AISdailyaggregates = pd.DataFrame(columns=["Installation ID","aggregate_date","hours_moving","hours_not_moving","hours_uncertain","completeness","distance_moved","min_speed","max_speed","avg_draught","avg_abs_turn_rate","record_count","last_position_time","last_position_latitude","last_position_longitude","last_speed"])
with open("/Users/alexeylevakin/Documents/SeaDevCon18/enrichment/SeaDevCon_AISdailyaggregates_sanitized.csv") as file:
    reader = csv.reader(file, delimiter=';', skipinitialspace=True)
    for r in reader:
        try:
            AISdailyaggregates = AISdailyaggregates.append({
                'Installation ID': r[0],
                'aggregate_date': r[1],
                'hours_moving': r[2],
                'hours_not_moving': r[3],
                'hours_uncertain': r[4],
                'completeness': r[5],
                'distance_moved': r[6],
                'min_speed': r[7],
                'max_speed': r[8],
                'avg_draught': r[9],
                'avg_abs_turn_rate': r[10],
                'record_count': r[11],
                'last_position_time': r[12],
                'last_position_latitude': r[13],
                'last_position_longitude': r[14],
                'last_speed': r[15]
            }, ignore_index = True)
        except BaseException as e:
            print(e)
```
